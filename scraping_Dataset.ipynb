```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

maxTweets = 5000
tweets_list = []
replies_list = []

# Search for the original tweet
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('jerawat lang:id until:2022-03-05 since:2020-03-01').get_items()):
    if i > maxTweets:
        break

    # Append the original tweet to the list
    tweets_list.append([
        tweet.date,
        tweet.id,
        tweet.content,
        tweet.user.username,
        tweet.user.displayname,
        tweet.user.location,
        tweet.user.followersCount,
        tweet.user.friendsCount,
        tweet.user.verified,
        tweet.retweetCount,
        tweet.lang,
        tweet.media,
        tweet.user.description
    ])

    # Check if the tweet is a reply and retrieve the username of the tweet being replied to
    if tweet.inReplyToUser:
        replies_list.append(tweet.inReplyToUser.username)
    else:
        replies_list.append(None)

# Create a DataFrame from the original tweets and replies
columns = ['Datetime', 'Tweet Id', 'Text', 'Username', 'DisplayName', 'Location', 'Followers', 'Following', 'Verified',
           'Retweet', 'Language', 'Media', 'Description']
tweets_df = pd.DataFrame(tweets_list, columns=columns)
tweets_df['Reply Username'] = replies_list

# Display the DataFrame
print(tweets_df)
```
